```python
# load and display an image with Matplotlib
import os
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt


from matplotlib.widgets import Slider
```

```python
# load image as pixel array
# An image of a file opens
image = mpimg.imread(os.path.join(os.getcwd(), "shely720x521.jpg")) / 255

plt.imshow(image)
plt.axis('off')  # Para desactivar los ejes en la visualización
plt.show()
```

```python
heigth, width, channels = image.shape
```

```python
modified_img = np.copy(image)
```

```python
for y in range(heigth):
    
    for x in range(width):
        
        r, g, b = modified_img[y, x]

        # Calcular componentes Y, I y Q
        change_y = 0.299 * r + 0.587 * g + 0.114 * b
        change_i = 0.595716 * r - 0.274453 * g - 0.321263 * b
        change_q = 0.211456 * r - 0.522591 * g + 0.311135 * b

        # Actualizar valores en la nueva matriz YIQ
        modified_img[y, x] = [change_y, change_i + 0.5, change_q + 0.5]        
```

# PDI – Mapas cromáticos

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# Generar datos de altitud de ejemplo
latitudes = np.linspace(-90, 90, 360)
longitudes = np.linspace(-180, 180, 720)
elevaciones = np.random.rand(len(latitudes), len(longitudes)) * 5000  # Altitudes en metros

# Crear una malla de latitudes y longitudes
X, Y = np.meshgrid(longitudes, latitudes)

# Crear la figura y los subgráficos en columnas
fig, axes = plt.subplots(3, 1, figsize=(8, 15), subplot_kw={'projection': '3d'})

# Visualizar el mapa de altitud con diferentes paletas
for i, cmap in enumerate(['gray', 'viridis', 'rainbow']):
    ax = axes[i]
    surf = ax.plot_surface(X, Y, elevaciones, cmap=cmap, linewidth=0, antialiased=False)
    ax.set_title(f'Paleta: {cmap.capitalize()}')
    ax.set_xlabel('Longitud')
    ax.set_ylabel('Latitud')
    ax.set_zlabel('Altitud (metros)')

    # Añadir barra de colores
    fig.colorbar(surf, ax=ax, label='Altitud (metros)')

plt.show()
```

# Se une todo el trabajo en una clase con atributos y funciones

```python
class ImageOperation:
    
    def __init__(self, img):
        
        self.img = img
    
    def rgbToYIQ(self):
        
        heigth, width, channels = self.img.shape
        
        modified_img = np.copy(self.img)
        
        for y in range(heigth):
    
            for x in range(width):

                r, g, b = modified_img[y, x]

                change_y = 0.299 * r + 0.587 * g + 0.114 * b
                change_i = 0.595716 * r - 0.274453 * g - 0.321263 * b
                change_q = 0.211456 * r - 0.522591 * g + 0.311135 * b

                modified_img[y, x] = [change_y, change_i + 0.5, change_q + 0.5]        
        
        return modified_img
    
    def yiqToRGB(self):
        
        heigth, width, channels = self.img.shape
        modified_img = np.copy(self.img)

        for a in range(heigth):

            for b in range(width):

                y, i, q = modified_img[a, b]

                i -= 0.5
                q -= 0.5

                change_r = y + 0.9663 * i + 0.6210 * q

                change_g = y - 0.2721 * i - 0.6474 * q

                change_b = y - 1.1070 * i + 1.7046 * q        

                modified_img[a, b] = [change_r, change_g, change_b]
        
        return modified_img
```

```python
imagen = mpimg.imread(os.path.join(os.getcwd(), "shely720x521.jpg")) / 255
```

```python
im = ImageOperation(imagen)
```
